### Prelimanary data cleaning

```python
# let's start with dropping our duplicates rows
df = df.drop_duplicates()

df.head(10)
```

```python
df.shape
```

- we dropped 29 duplicated row

```python
# lowering all columns names into small case letters
df.columns = map(str.lower, df.columns)

df.head(3)
```

```python
# one of our columns have a wrong name lets rename it:
df = df.rename({'unnamed: 0':'ID' }, axis=1)
df.head(3)
```

```python
# lets then check our datatypes
df.dtypes
```

```python
# since we have float and object datatypes its informative to check nunique()
df.nunique()
```
